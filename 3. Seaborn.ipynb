# 2. which are the cities which hosted oymics?which city hosted maximum number of olympics?

```python

host=athl.groupby(['City','Year']).size()
host=pd.DataFrame(host)
host=host.reset_index()
host=host.rename(columns={0:'Count'})
host
```

# 3. Top 10 team having maximum no of athlete

```python
h=athl.groupby(['Team','Name']).size()
h=pd.DataFrame(h)
h=h.reset_index()
h=h.rename(columns={0:'Count'})
h
```

```python
h1=h.groupby(['Team']).size()
h1=pd.DataFrame(h1)
h1=h1.reset_index()
h1=h1.rename(columns={0:'Count'})
h1=h1.sort_values(by='Count',ascending=False)
h1=h1.head(10)
h1
```

```python
fig=plt.figure(figsize=(12,7))
plt.bar(h1['Team'],h1['Count'])
plt.xticks(rotation=90)
plt.show()
```

# 4. top 10 team having maximum avearage height of athletes

```python
a=athl[['Team','Height']]
a=a.groupby(['Team']).mean()
a=pd.DataFrame(a)
a=a.reset_index()
a=a.sort_values(by='Height',ascending=False)
a=a.head(10)
a
```

```python
fig=plt.figure(figsize=(7,5))
plt.bar(a['Team'],a['Height'])
plt.xticks(rotation=90)
plt.show()
```

# 5.top 10 team having maximum avearage Weight of athletes

```python
b=athl[['Team','Weight']].groupby(['Team']).mean()
b=pd.DataFrame(b)
b=b.reset_index()
b=b.sort_values(by='Weight',ascending=False)
b=b.head(10)
b
```

```python
fig=plt.figure(figsize=(7,5))
plt.bar(b['Team'],b['Weight'],color='red')
plt.xticks(rotation=90)
plt.show()
```

# 6. top 10 sports that have maximum participation

```python
import seaborn as sns
import matplotlib.pyplot as plt
import  numpy as nm
import pandas as pd
import matplotlib as mpl
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

# bar and line

```python
sns.distplot(tips['total_bill'],color='r')
mpl.rcParams['patch.force_edgecolor']=True
#once only after that below command for all will work
mpl.rcParams['patch.edgecolor']='blue'
```

```python
#bar with line chart
plt.figure(figsize=(10,7))
sns.distplot(tips['tip'],kde=True)#kde is the line true for line show
#false for hide line
plt.title('distribution of tips')
plt.xlabel('Tips')
plt.ylabel('Frequency')
plt.show()
```

# jointplots

```python
sns.jointplot(x='total_bill',y='tip',data=tips,color='teal',size=9)
sns.jointplot(x='total_bill',y='tip',color='r',data=tips,kind='hex',ratio=1)
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg',color='khaki')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid',color='forestgreen')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde',color='pink')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter',color='darkseagreen')
```

# pair plot is used for comparing each numerical column 

```python
sns.pairplot(tips)
sns.pairplot(tips,hue='sex',palette='cool')
sns.pairplot(tips,hue='sex',palette='hsv',kind='reg')
sns.pairplot(tips,hue='sex',palette='cool',kind='scatter')
#for line use diag_kind
sns.pairplot(tips,hue='sex',palette='cool',diag_kind='kde')
```

# drawind line chart

```python
sns.kdeplot(tips['tip'])
sns.kdeplot(tips['tip'],shade=True)
sns.kdeplot(tips['tip'],shade=True,legend=True)
```

# working on iris

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
grid=sns.PairGrid(iris)
grid.map(plt.scatter)
plt.show()
```

```python
grid=sns.PairGrid(iris,hue='species',palette='magma_r')
grid.map(plt.scatter)
plt.show()
```

```python
grid=sns.PairGrid(iris)
grid.map_diag(plt.hist)
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot)
plt.show()
```

```python
grid=sns.PairGrid(iris)
grid.map_diag(plt.hist,color='r')
grid.map_upper(plt.scatter,color='blue')
grid.map_lower(sns.kdeplot)
plt.show()
```

```python
grid=sns.PairGrid(iris,hue='species',palette='hot')
grid.map_diag(plt.hist)
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot)
plt.show()
```
